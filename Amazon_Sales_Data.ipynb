# INTRODUCTION
### This dataset contains the data of more than 1000 Amazon Product's Ratings and Reviews as per their details listed on the official website of Amazon. The objective of this project is to clean the data to perform further analysis and create a visualization

```python
# cleaning the discounted price column and also changing its data type from object to float
Amazon_Sales_Data['discounted_price'] = Amazon_Sales_Data['discounted_price'].str.replace("₹",'')
Amazon_Sales_Data['discounted_price'] = Amazon_Sales_Data['discounted_price'].str.replace(",",'')
Amazon_Sales_Data['discounted_price'] = Amazon_Sales_Data['discounted_price'].astype('float64')
```

```python
# cleaning the actual price column and also changing its data type from object to float
Amazon_Sales_Data['actual_price'] = Amazon_Sales_Data['actual_price'].str.replace("₹",'')
Amazon_Sales_Data['actual_price'] = Amazon_Sales_Data['actual_price'].str.replace(",",'')
Amazon_Sales_Data['actual_price'] = Amazon_Sales_Data['actual_price'].astype('float64')
```

## Performing Exploratory data analysis (EDA)

```python
#plotting a bar graph to show 5 most cheapest products after discount
# Define a list of colors
colors = ['red', 'green', 'blue', 'orange', 'purple']

# Sort the data and select the top 5 rows
df_sorted = Amazon_Sales_Data.sort_values('Price_difference', ascending=False).head(5)

# Create a horizontal bar plot
fig, ax = plt.subplots()
ax.barh(df_sorted['product_name'], df_sorted['Price_difference'], color=colors)

# Customize the plot
ax.set_title('Top 5 Cheapest Products (Discounted)', fontweight='bold')
ax.set_xlabel('Price after Discount (IND Rupees)', fontweight='bold')
ax.set_ylabel('Product Name', fontweight='bold')

plt.show()
```

```python
#plotting a bar graph to show 5 most expensive products after discount
# Define a list of colors
colors = ['purple', 'green', 'blue', 'orange','red' ]

# Sort the data and select the top 5 rows
df_sorted = Amazon_Sales_Data.sort_values('Price_difference', ascending=False).head(5).iloc[::-1]

# Create a horizontal bar plot
fig, ax = plt.subplots()
ax.barh(df_sorted['product_name'], df_sorted['Price_difference'], color=colors)

# Customize the plot
ax.set_title('Top 5 Expensive Products (Discounted)', fontweight='bold')
ax.set_xlabel('Price after Discount (IND Rupees)', fontweight='bold')
ax.set_ylabel('Product Name', fontweight='bold')

plt.show()
```

```python
#plotting a graph to show the major products by Primary Category and Secondary Category
fig, ax = plt.subplots(2, 1, figsize=(8, 10))
fig.suptitle('Major Products by Category', fontweight='heavy', size='x-large')

main_colors = sns.color_palette('colorblind', n_colors=5)
sub_colors = sns.color_palette('bright', n_colors=10)

(Amazon_Sales_Data['Primary_Category'].value_counts()
    .head(5)
    .rename_axis('Primary_Category')
    .reset_index(name='counts')
    .pipe((sns.barplot, 'data'), ax=ax[0], x='counts', y='Primary_Category', palette=main_colors))

(Amazon_Sales_Data['Secondary_Category'].value_counts()
    .head(10)
    .rename_axis('Secondary_Category')
    .reset_index(name='counts')
    .pipe((sns.barplot, 'data'), ax=ax[1], x='counts', y='Secondary_Category', palette=sub_colors))

plt.subplots_adjust(hspace=0.3)

for i in range(2):
    ax[i].set_xlabel('Count', fontweight='bold')
    ax[i].set_ylabel(f'Product by {"Primary " if i==0 else "Secondary "}Category', fontweight='bold')
    ax[i].set_title(f"Most Products by {'Primary ' if i==0 else 'Secondary '}Category", fontweight='bold')
    ax[i].bar_label(ax[i].containers[0])

plt.show()
```

```python
# plotting a scattered graph to find relation between actual price and discounted price
x = Amazon_Sales_Data['actual_price']
y = Amazon_Sales_Data['discounted_price']
plt.figure(figsize=(15,8))
colors = Amazon_Sales_Data['Primary_Category'] # assuming Category is a categorical variable

plt.scatter(x, y, c='purple')
plt.title('Relation Between Actual Price vs Discounted Price', fontweight='bold')
plt.xlabel('Actual Price', fontweight='bold')
plt.ylabel('Discounted Price', fontweight='bold')
plt.colorbar() # adds a colorbar legend
plt.show()
```
